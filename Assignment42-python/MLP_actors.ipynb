```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import wandb
from wandb.keras import WandbCallback
```

```python
wandb.login()
```

```python
wandb.init(project="deepface", entity="gharabadiyan")
```

```python
config = wandb.config
config.learning_rate = 0.001
```

```python
data = pd.read_csv('Arcface_data.csv')
data
```

```python
X = data.iloc[:, 0:512].values
Y = data[['label']]

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2 ,shuffle=True)
```

## Model

```python
model = tf.keras.models.Sequential([
    Dense(300,input_dim=512,activation="relu"),
    Dense(128,activation="relu"),
    Dense(64,activation="relu"),
    Dense(3,activation="softmax")
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=config.learning_rate),
              loss = tf.losses.sparse_categorical_crossentropy,
              metrics=['accuracy'])
```

```python
model.fit(X_train,Y_train,epochs=150,callbacks=[WandbCallback()])
```

```python
model.evaluate(X_test,Y_test)
```
